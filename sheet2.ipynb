```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

# problem 8

```python
x = np.array([7.3, 18.4, 27.6, 42.1, 57.4]).reshape((-1, 1)) # g/dm^3
y = np.array([0.211,0.533,0.804,1.236,1.701])  # kPa
#plt.plot(x,y, '-x')
plt.scatter(x,y)
plt.xlabel('Protein Concentration (in g/dm^3 )')
plt.ylabel('Osmotic Pressure (in kPa)')
plt.show()
```

```python
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
R = 8.314 # kPa dm3 mol-1 K-1
T = 21 +  273.15 # K
g_per_mol = R*T/model.coef_[0]
print(g_per_mol, 'g_per_mol')
```

# Problem 9

```python
xdata = np.array([1.6,2.4,3.2,4.1])
ydata = np.array([3.32,5.3,9.05,16.3])
#plt.plot(xdata,ydata,'-x')
plt.scatter(xdata,ydata)
plt.xlabel('x')
plt.ylabel('y')
```

```python
def func(x,A,B):
    return A*x*x*x+B
    
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
plt.scatter(xdata,ydata, label='data')
plt.plot(xdata, func(xdata, *popt), 'r:',

         label='fit: A=%5.3f, B=%5.3f' % tuple(popt))
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
