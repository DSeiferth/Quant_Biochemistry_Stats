# Stats sheet 1 - distributions

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
from scipy.special import factorial
f_size = 22
```

# binomial distribution

```python
def binom_distr(k,n,p):
    '''
    inputs:
    outputs:
    '''
    return factorial(n)/(factorial(k)*factorial(n-k))*np.power(p,k)*np.power(1-p,n-k)
    
```

```python
k = np.arange(0,7,1)
distr = binom_distr(k=k,n=6,p=0.4)
print(distr)
```

```python
fig, ax = plt.subplots()
plt.title('Binomial distribution', fontsize=f_size)
plt.scatter(k,distr, label='data')
plt.xlabel('k', fontsize=f_size)
plt.ylabel('Prob', fontsize=f_size)
ax.tick_params(axis='both', which='major', labelsize=f_size)
plt.show()
```

# Poisson distribution

```python
def poisson_distr(x, lam):
    return np.power(lam, x)*np.exp(-lam)/factorial(x)
```

```python
x = np.arange(0,10,1)
distr = poisson_distr(x=x,lam=1.8)
print(distr)
```

```python
fig, ax = plt.subplots()
plt.title('Poisson distribution', fontsize=f_size)
plt.scatter(x,distr, label='data')
plt.xlabel('k', fontsize=f_size)
plt.ylabel('Prob', fontsize=f_size)
ax.tick_params(axis='both', which='major', labelsize=f_size)
plt.show()
```
